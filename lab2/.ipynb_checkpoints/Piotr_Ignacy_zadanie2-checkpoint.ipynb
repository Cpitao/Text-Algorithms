### Zapewnione funkcje do rysowania grafow wraz z klasą reprezentującą węzeł

```python
import tempfile
import subprocess
from IPython.display import Image, display

def view_pydot(pdot):
    plt = Image(pdot.create_png())
    display(plt)

def show_pydot(description):
    with tempfile.NamedTemporaryFile(mode="w") as tmp1:
        tmp1.write(description)
        tmp1.flush()
        bashCommand = f"dot -Tpng {tmp1.name}"
        process = subprocess.Popen(bashCommand.split(), stdout=subprocess.PIPE)
        output, error = process.communicate()
        display(Image(output))
```

```python
class Node:
    counter = 1
    
    def __init__(self, label, depth):
        self.children = {}
        self.label = label
        self.link = None
        self.id = Node.counter
        self.depth = depth
        self.interval = None
        Node.counter += 1
 
    def dot_node(self):
        return f'{self.id} [label="", shape="point"];'
        
    def dot_edges(self, links=True):
        result = []
        for child in self.children.values():
            result.append(f'{self.id} -> {child.id}[arrowhead="none", label=" {child.label}"];')
        if self.link and links:
            result.append(f'{self.id} -> {self.link.id}[color="blue"];')
        return "\n".join(result)
    
    def all_children(self, only_leafs=False):
        if(not only_leafs or not self.children):
            result = [self]
        else:
            result = []
        for child in self.children.values():
            result += child.all_children(only_leafs=only_leafs)
        return result
    
    def show(self, links=True):
        nodes = self.all_children()
        Node.show_graph(nodes, links=links)
    
    @staticmethod
    def show_graph(nodes, links=True):
        result = "digraph{\n"
        ranks = {}
        for node in nodes:
            result += node.dot_node() + "\n"
            result += node.dot_edges(links) + "\n"
            if(node.depth not in ranks):
                ranks[node.depth] = []
            ranks[node.depth].append(node.id)
        for key, values in ranks.items():
            values = " ".join([f"{v};" for v in values])
            result += " { rank=same; " + values + "}\n"
        result += "\n}"
        ############## MODYFIKACJA DLA WINDOWSA ##############
        ############## NA LINUXIE NALEŻY ODKOMENTOWAĆ PONIŻSZĄ LINIĘ I ZAKOMENTOWAĆ 2 PONIŻEJ ##############
#         show_pydot(result)
        s = Source(result)
        display(s)
```

### Drzewo Trie

```python
def init_trie(text):
    root = Node("", 0)
    current = root
    for c in text:
        current.children[c] = Node(c, current.depth+1)
        current = current.children[c]
    
    return root

def add_tree_part(root, text, start):
    current = root
    for i in range(start, len(text)):
        current.children[text[i]] = Node(text[i], current.depth+1)
        current = current.children[text[i]]

def get_head(node, suffix, start):
    current = node
    for i in range(start, len(suffix)):
        if suffix[i] in current.children: # O(1) ze względu na sposób w jaki działa słownik
            current = current.children[suffix[i]]
        else:
            break

    return current

def build_trie(text):
    trie = init_trie(text)
    for i in range(1, len(text)):
        head = get_head(trie, text, i) # przekazujac trzeci parametr nie musimy kopiowac obiektu
        add_tree_part(head, text, i+head.depth)
    
    return trie

def contains(trie, word):
    for c in word:
        if c in trie.children:
            trie = trie.children[c]
        else:
            return False
    return True
```

```python
for d in data:
    trie = build_trie(d)
    trie.show()
    print("Contains aab:", "trie says", contains(trie, "aab"), "and find says", (lambda x: False if x == -1 else True)(d.find("aab")))
    print("Contains aba:", "trie says", contains(trie, "aba"), "and find says", (lambda x: False if x == -1 else True)(d.find("aba")))
    print("Contains bbb:", "trie says", contains(trie, "bbb"), "and find says", (lambda x: False if x == -1 else True)(d.find("bbb")))
    print("Contains bcb:", "trie says", contains(trie, "bcb"), "and find says", (lambda x: False if x == -1 else True)(d.find("bcb")))
```

### Drzewo suffixowe

```python
def slow_suffix_head(root, text, start):
    current = root
    i = start
    while i < len(text):
        if text[i] in current.children:
            if current.children[text[i]].interval is not None: # jeśli krawędź do danego dziecka oznaczała przedział
                start, end = current.children[text[i]].interval
                if text[start:end] != text[i:min(i+end-start, len(text))]: # jeśli cały przedział się nie pokrywa zwróć węzeł
                    return current.children[text[i]]
                else: # jeśli się pokrywa - idź dalej w dół drzewa
                    current = current.children[text[i]]
                    i += end - start
            else: # jeśli dana krawędź oznacza pojedynczy znak - podążamy w dół drzewa po tej krawędzi
                current = current.children[text[i]]
                i += 1
        else: # jeśli kolejna szukana litera nie jest dzieckiem w danym węźle - zwróć aktualny węzeł
            return current

def insert_suffix(root, head, text, start):
    if head.interval is not None: # jeśli dana krawędź reprezentuje przedział
        i = 1
        while start + i + head.depth - 1 < len(text) and i < head.interval[1] - head.interval[0] and \
              text[head.interval[0] + i] == text[start + head.depth + i - 1]:

            i += 1
        # split interval into two - (interval[0], interval[0] + i) and (interval[0] + i, interval[1])
        first_interval = (head.interval[0], head.interval[0] + i)
        second_interval = (head.interval[0] + i, head.interval[1])
        if first_interval[1] - first_interval[0] == 1:
            head.label = text[first_interval[0]]
            head.interval = None
        else:
            head.label = f"[{first_interval[0]}, {first_interval[1]}]"
            head.interval = first_interval
        
        # appending previously existing part
        if second_interval[1] - second_interval[0] == 1:
            old_node = Node(text[second_interval[0]], head.depth + first_interval[1] - first_interval[0]) # +1?
        else:
            old_node = Node(f"[{second_interval[0]}, {second_interval[1]}]",
                            head.depth + first_interval[1] - first_interval[0]) # +1?
            old_node.interval = second_interval
        old_node.children = head.children
        head.children = {}
        head.children[text[second_interval[0]]] = old_node
        
        # appending new node
        if start + head.depth + first_interval[1] - first_interval[0] == len(text):
            new_node = Node(text[-1], head.depth + first_interval[1] - first_interval[0]) # +1?
        else:
            new_node = Node(f"[{start + head.depth + first_interval[1] - first_interval[0] - 1}, {len(text)}]", 
                            head.depth + first_interval[1] - first_interval[0]) # +1?
            new_node.interval = (start + head.depth + first_interval[1] - first_interval[0] - 1, len(text))
        head.children[text[start + head.depth + first_interval[1] - first_interval[0] - 1]] = new_node
    else:
        if start + head.depth == len(text) - 1:
            new_child = Node(text[start + head.depth], head.depth + 1)
        else:
            new_child = Node(f"[{start + head.depth}, {len(text)}]", head.depth + 1)
            new_child.interval = (start + head.depth, len(text))
        head.children[text[start + head.depth]] = new_child


def build_suffix_tree(text, draw=True):
    root = Node("", 0)
    for i in range(len(text)):
        insert_suffix(root, slow_suffix_head(root, text, i), text, i)
        if draw:
            root.show()
    
    return root

def contains_suffix(root, word, text):
    i = 0
    while i < len(word):
        if root.interval is not None:
            for j in range(root.interval[0] + 1, root.interval[1]):
                if i < len(word) and word[i] == text[j]:
                    i += 1
                elif i == len(word):
                    return True
                else:
                    return False
        if i == len(word):
            return True
        else:
            if word[i] in root.children:
                root = root.children[word[i]]
                i += 1
            else:
                return False
    return True
            
```

```python
for d in data:
    st = build_suffix_tree(d, draw=False)
    st.show()
    print("Contains aab:", "suffix tree says", contains_suffix(st, "aab", d), "and find says",
          (lambda x: False if x == -1 else True)(d.find("aab")))
    print("Contains aba:", "suffix tree says", contains_suffix(st, "aba", d), "and find says",
          (lambda x: False if x == -1 else True)(d.find("aba")))
    print("Contains bbb:", "suffix tree says", contains_suffix(st, "bbb", d), "and find says",
          (lambda x: False if x == -1 else True)(d.find("bbb")))
    print("Contains bcb:", "suffix tree says", contains_suffix(st, "bcb", d), "and find says",
          (lambda x: False if x == -1 else True)(d.find("bcb")))
```

## Testy szybkości

```python
import matplotlib.pyplot as plt
import time

results = [[], []]
test_data = ["bbb$", "aabbabd", "ababcd", "abcbccd"]
with open("1997_714_head.txt", "r", encoding='utf-8') as f:
    text = f.read()
text += "&" # unikalny znak na końcu tekstu
test_data.append(text)
for d in test_data:
    start = time.time()
    build_trie(d)
    end = time.time()
    results[0].append(end - start)
    
    start = time.time()
    build_suffix_tree(d, draw=False)
    end = time.time()
    results[1].append(end - start)


labels = ["bbb$", "aabbabd", "ababcd", "abcbccd", "given text"]
_, ax = plt.subplots()
ax.scatter(labels, results[0], label="Trie")
ax.scatter(labels, results[1], label="Suffix tree")

ax.legend()
ax.set_xlabel("Długość stringa")
ax.set_ylabel("Czas wykonania [s]")
plt.show()
```
